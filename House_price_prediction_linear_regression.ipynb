```python
plt.style.use("seaborn-v0_8")
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color ="Red",marker='+')
```

```python
X = df[['area']] #independent
y = df[['price']] #dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
y_train
```

```python
X_train
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

$$y = mx +c$$
$$price = m(area) + c$$

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X_train,y_train) #the model learns the pattern
```

```python
X_test
```

```python
y_test
```

```python
predicted_values = lin.predict(X_test)
predicted_values   #we gave X_test to predict on these data
```

```python
lin.coef_  #slope - value of m
```

```python
lin.intercept_  #intercept - value of c
```

```python
lin.score(X_train,y_train)  #training efficiency-- Learning effeiciency
```

```python
lin.score(X_test,y_test) #model efficiency for prediction
```

```python
error = y_test-predicted_values
error
```
